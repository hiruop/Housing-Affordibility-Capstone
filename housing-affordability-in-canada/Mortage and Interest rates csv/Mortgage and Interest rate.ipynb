```python
# mortgage rate data, changed column names and dropped empty columns
mortgageData = pd.read_csv('C:/Users/Hiren/Desktop/ST.Clair/sem 4/402-Capstone/housing-affordability-in-canada/Mortage and Interest rates csv/34100145.csv')
mortgageData = mortgageData.drop(columns=['GEO', 'DGUID', 'UOM', 'UOM_ID', 'SCALAR_FACTOR', 'SCALAR_ID', 'VECTOR', 'COORDINATE', 'STATUS', 'SYMBOL', 'TERMINATED', 'DECIMALS'])
mortgageData = mortgageData.rename(columns={'REF_DATE': 'Date', 'VALUE': 'Mortgage Rate'})
```

```python
mortgageData.head()
```

```python
# interest rate data, changed column names
interestData = pd.read_csv('C:/Users/Hiren/Desktop/ST.Clair/sem 4/402-Capstone/housing-affordability-in-canada/Mortage and Interest rates csv/IRSTCB01CAM156N.csv')
interestData = interestData.rename(columns={'DATE': 'Date', 'IRSTCB01CAM156N': 'Interest Rate'})
```

```python
interestData.tail()
```

```python
# merged dataframes together and aligned by date
rates = pd.concat([mortgageData, interestData])
# trying to convert year data to integers to remove decimal values
#finalAllData['Year'] = finalAllData['Year'].dropna().astype(int)
rates = rates.reset_index(drop=True)
#finalAllData = finalAllData.rename(columns = {'Unnamed 0':'Year'})
aggFunc = {'Mortgage Rate': 'sum', 'Interest Rate': 'sum'}
rates = rates.groupby(rates['Date']).aggregate(aggFunc)
rates.head(7)
```

```python
rates.to_csv('out.csv', index=False)
```
